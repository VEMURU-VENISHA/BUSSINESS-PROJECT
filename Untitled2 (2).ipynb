```python
variables = ["PPU"]
categories = ["Region"]

for var in variables:
    for cat in categories:
        print(f"\n=== Mean {var} by {cat} ===")
        print(df.groupby(cat)[var].agg(['mean', 'std', 'count']).sort_values(by='mean', ascending=False))
```

```python
for var in variables:
    for cat in categories:
        print(f"\nFull description of {var} by {cat}")
        print(df.groupby(cat)[var].describe())
```

```python
variables = ["PPU"]
categories = ["Region"]
beverages = df["Beverage_Brand"].unique()

for bev in beverages:
    print(f" Statistics for {bev}")


    df_bev = df[df["Beverage_Brand"] == bev]

    for var in variables:
        for cat in categories:
            print(f"\n--- {var} by {cat} ---")
            stats = df_bev.groupby(cat)[var].agg(['mean', 'std', 'count', 'min', 'max']).sort_values(by='mean', ascending=False)
            print(stats)
```

```python
anova_retailer_OM = smf.ols('OM ~ C(Retailer)', data=df).fit()
print("\nOne-Way ANOVA (Retailer on OM):\n", sm.stats.anova_lm(anova_retailer_OM, typ=2))
```

```python
anova_region_OM = smf.ols('OM ~ C(Region)', data=df).fit()
print("\nOne-Way ANOVA (Region on OM):\n", sm.stats.anova_lm(anova_region_OM, typ=2))
```

```python
anova_beverage_OM = smf.ols('OM ~ C(Beverage_Brand)', data=df).fit()
print("\nOne-Way ANOVA (Beverage on OM):\n", sm.stats.anova_lm(anova_beverage_OM, typ=2))
```

```python
anova_state_OM = smf.ols('OM ~ C(State)', data=df).fit()
print("\nOne-Way ANOVA (State on OM):\n", sm.stats.anova_lm(anova_state_OM, typ=2))
```

```python
!pip install pingouin

import pingouin as pg

variables = ["OM", "OP", "TS"]
categories = ["Retailer", "Region", "Beverage_Brand", "State"]

for var in variables:
    for cat in categories:
        print(f"\n=== GamesHowell test for {var} by {cat} ===")
        result = pg.pairwise_gameshowell(dv=var, between=cat, data=df)
        print(result)
```

```python
from matplotlib.ticker import FuncFormatter
data = df[['Beverage_Brand', 'Region', 'TS', 'OP']]
data = data.dropna()

grouped = data.groupby(['Beverage_Brand', 'Region']).agg(
    Avg_Total_Sales=('TS', 'mean'),
    Avg_Operating_Profit=('OP', 'mean')
).reset_index()

sales_pivot = grouped.pivot(index='Beverage_Brand', columns='Region', values='Avg_Total_Sales').fillna(0)
profit_pivot = grouped.pivot(index='Beverage_Brand', columns='Region', values='Avg_Operating_Profit').fillna(0)


plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'figure.facecolor': 'white', 'axes.facecolor': 'white'})


fig, ax = plt.subplots(figsize=(12,6))
sales_pivot.plot(kind='bar', ax=ax)
ax.set_title('Average Total Sales by Beverage Brand and Region', fontsize=16, weight='bold')
ax.set_xlabel('Beverage Brand', fontsize=12)
ax.set_ylabel('Average Total Sales', fontsize=12)
ax.legend(title='Region', bbox_to_anchor=(1.02, 1), loc='upper left')
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.show()
fig, ax = plt.subplots(figsize=(12,6))
profit_pivot.plot(kind='bar', ax=ax, colormap='Blues')
ax.set_title('Average Operating Profit by Beverage Brand and Region', fontsize=16, weight='bold')
ax.set_xlabel('Beverage Brand', fontsize=12)
ax.set_ylabel('Average Operating Profit', fontsize=12)
ax.legend(title='Region', bbox_to_anchor=(1.02, 1), loc='upper left')
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(sales_pivot, cmap='Blues', annot=True, fmt=".0f", linewidths=.5, cbar_kws={'label': 'Avg Total Sales'})
plt.title('Heatmap: Average Total Sales by Brand and Region', fontsize=16, weight='bold')
plt.xlabel('Region')
plt.ylabel('Beverage Brand')
plt.tight_layout()
plt.show()
```
